Імпортування бібліотек:

```python
import pandas as pd
import numpy as np
```

Зчитуємо файл:

```python
data = pd.read_csv("cities.csv")
```

Ознайомимось із датасетом:

```python
data.head()
```

Зробимо необхідний препроцесінг даних:

```python
data.columns = data.columns.str.replace('"', '')
data.iloc[:, -2] = data.iloc[:, -2].str.lower()
data.iloc[:, -2] = data.iloc[:, -2].str.replace('"', '')
data.iloc[:, -2] = data.iloc[:, -2].str.replace(' ', '')
data.head()
```

Введіть назву бажаного міста зі списку, що надається при виводі. Програма обраховує jaccard similarity між даним містом та наступним містом. Якщо введенего міста немає, то виводиться помилка та прохання вибрати інше місто. Якщо вводиться останнє місто, то виводиться jaccard similarity між ним та першим містом.

```python
print("cities: ", list(data.iloc[:, -2]))

# just choose a city
# returns jaccard similarity between your city and the next one

print("Enter the name of city (copy from list above)")
city_1 = input()

try:
  print(data.loc[data.iloc[:, -2] == city_1].index[0])
except:
  print("choose another one, it is impossible")

nxt = data.loc[data.iloc[:, -2] == city_1].index[0] + 1

try:
  city_2 = data.iloc[:, -2][nxt]
  el_1 = set(city_1)
  el_2 = set(city_2)
  intersection = el_1.intersection(el_2)  
  union = el_1.union(el_2)
  jaccard = len(intersection) / len(union)
  print(f"Similarity between {city_1} and {city_2}:", jaccard)
except:
  city_1 = set('ravenna')
  city_2 = set('youngstown')
  intersection = city_1.intersection(city_2)  
  union = city_1.union(city_2)
  jaccard = len(intersection) / len(union)
  print(f"Similarity between ravenna and youngstown:", jaccard)
```
